# i)Generating random samples from the poisson distribution

```python
from numpy import random
```

```python
x = random.poisson(lam=5,size=1000)
print(x)
```

## ii)Calculating mean and variance

### Mean

```python
import numpy as np
```

```python
mean = np.mean(x)
print(mean)
```

### Variance

```python
Variance = np.var(x)
print(Variance)
```

# iii)Plotting a histogram

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from numpy import random 
```

```python
poisson_data = random.poisson(lam =5,size = 1000)# Lam represents the number of occurencies
sns.histplot(poisson_data,label = 'poisson',kde = 'False',color ='violet', edgecolor = 'purple' )# kde helps in visualisation of data
plt.xlabel('Customers per day', color ='purple')
plt.ylabel('Generated Samples',color='purple')
plt.title('A HISTOGRAM OF THE SAMPLES FROM THE POISSON DISTRIBUTION',color='indigo')
plt.show()
```
